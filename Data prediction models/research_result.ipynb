```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = 20, 10

model = ['ARIMA', 'LSTM', 'XGBoost', 'White noise', 'Blue noise']
RMSE = [0.488, 1.613, 5.346, 3.001, 2.796]
data = pd.DataFrame({'model' : model, 'RMSE' : RMSE})
sns.barplot(data=data, x='model', y='RMSE')
plt.xlabel("Model", fontsize = 25)
plt.ylabel("RMSE", fontsize = 25)
plt.title("Model's RMSE comparison", fontsize = 30)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```
